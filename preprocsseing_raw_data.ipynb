```python
import pandas as pd
import json
import os
```

```python
def concatenate_speaker_text(group):
    eojeol_sum = 0
    result = []
    prev_speaker = None
    combined_text1 = []
    combined_text2 = []
    
    for _, row in group.iterrows():
        if row['speaker_id'] != prev_speaker:
            if combined_text1 and combined_text2:
                result.append((prev_speaker, ' '.join(combined_text1), ' '.join(combined_text2), eojeol_sum))
            prev_speaker = row['speaker_id']
            eojeol_sum = row["eojeolList_cnt"]
            combined_text1 = [row['standard_form']]
            combined_text2 = [row['dialect_form']]
        else:
            eojeol_sum += row["eojeolList_cnt"]
            combined_text1.append(row['standard_form'])
            combined_text2.append(row['dialect_form'])
    
    if combined_text1 and combined_text2:
        result.append((prev_speaker, ' '.join(combined_text1), ' '.join(combined_text2), eojeol_sum))
    
    return pd.DataFrame(result, columns=['speaker_id', 'standard_form', 'dialect_form', "eojeol_sum"])


def is_eojeol_than_2(seq):
    result = 0
    for eojeol in seq:
        if eojeol["isDialect"] == True:
            result += 1
    return result
```

```python
# 결과를 저장할 빈 데이터프레임 생성
result = pd.DataFrame(columns=['id', 'form', 'standard_form', 'dialect_form', 'speaker_id', 'start', 'end', 'note', 'eojeolList'])

# 결과를 일단 리스트로 저장
data_list = []

file_path = '/Users/honey/Desktop/2024-2 KUBIG/contest/dataset/raw/경상도_학습데이터_통합본/DKCI'  # 파일 경로 설정

for i in range(3193):
    file_num = 20000000 + i
    file_name = file_path + str(file_num) + ".json"
    print(f"Processing file: {file_name}")

    try:
        with open(file_name, 'r', encoding='utf-8') as file:
            print("File opened successfully")
            json_data = json.load(file)
            utterance_data = pd.DataFrame(json_data["utterance"])
            speaker_data = pd.DataFrame(json_data["speaker"])

            result_data = pd.DataFrame()

            if len(speaker_data) == 2:
                utterance_data["eojeolList_cnt"] = utterance_data["eojeolList"].apply(lambda x : is_eojeol_than_2(x))
                result_data = concatenate_speaker_text(utterance_data)
                result_data["id"] = json_data["id"]

                # 마지막 행의 특정 칼럼 (예: 'B') 값이 True 인 경우 삭제
                if result_data.iloc[0]["speaker_id"] == '1':
                    if result_data.iloc[-1]['speaker_id'] == '1':
                        result_data = result_data.drop(result_data.index[-1])
                else:
                    if result_data.iloc[0]["speaker_id"] != result_data.iloc[-1]['speaker_id']:
                        result_data = result_data.drop(result_data.index[-1])

                grouped = list(result_data.groupby("speaker_id").count()["standard_form"])
                
                if grouped[0] == grouped[1]:
                    data_list.append(result_data)
                
    except FileNotFoundError:
        print(f"File not found: {file_name}")
        continue
    except json.JSONDecodeError:
        print(f"Error decoding JSON in file: {file_name}")
        continue
    except Exception as e:
        print(f"Unexpected error with file {file_name}: {e}")
        continue

# 리스트에 저장된 데이터프레임들을 한 번에 병합
if data_list:
    result_1 = pd.concat(data_list, ignore_index=True)

print("Data processing complete")
```
